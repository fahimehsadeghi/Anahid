With the following code example, we read each file row and split with "," each item of the row and put each one in a list. In total, we create 20 list. because there are 20 items in each row

```python
import csv
from collections import defaultdict
columns = defaultdict(list)
with open('C:...input\\data.csv', 'r') as f:
    reader = csv.reader(f, delimiter=',')
    for row in reader:
        for i in row:
            sample_data =i.split(',')
            
            for j in range(len(sample_data)):
                columns[j].append(sample_data[j])
```

Second, we convert the dictionary created in the previous step into a data frame

```python
data = {'Episode': columns.get(1), 'Station': columns.get(2),'Channel Type': columns.get(3), 'Season': columns.get(4),'Year': columns.get(5), 'Date': columns.get(6),'Day of week': columns.get(7), 'Start_time': columns.get(8),'End_time': columns.get(9), 'Length': columns.get(10),'Name of show': columns.get(11), 'Name of episode': columns.get(12),'Genre': columns.get(13), 'First time or rerun': columns.get(14),'# of episode in the season': columns.get(15), 'Movie?': columns.get(16),'Game of the Canadiens during episode?': columns.get(17),'Market Share_total': columns.get(18),'Temperature in Montreal during episode': columns.get(19)}
```

```python
import pandas as pd
dataset =pd.DataFrame.from_dict(data,orient='index')
```

Since the target column contains non-number irrelevant values, we have converted the values to null values with the following code example

```python
df =dataset.transpose()
df['Market Share_total'] = df['Market Share_total'].str.replace(r'\D', '')
df
```

We will then remove the null values created in the previous step

```python
df = df[df['Market Share_total'].str.strip().astype(bool)]
```

Since our classification algorithm is executed with numeric values, we have encoded all non-numeric values into unique numbers in the following code example

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df = df.apply(lambda col: le.fit_transform(col.astype(str)), axis=0, result_type='expand')
df
```

Since the classes in the target column are unbalanced, with the sample code below we have selected 50 samples from each class we have 500 different classes that will be created in the dataframe with 25000 rows. There are different ways to balance classes, for example, smote, but since the dataset is large, if you use one of these methods you will need a lot of RAM to process

```python
balanced_df=df.groupby('Market Share_total',as_index = False,group_keys=False).apply(lambda s: s.sample(50,replace=True))
balanced_df
```

Next we select the column 'Market Share_total' as the target column

```python
df_feature=balanced_df.loc[:, balanced_df.columns != 'Market Share_total'].values
df_labels =  balanced_df['Market Share_total'].values
```

Here we use the chi2 method to Feature Selection

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

chi2_selector = SelectKBest(chi2, k=10)
X_kbest = chi2_selector.fit_transform(df_feature, df_labels)
print(chi2_selector.scores_)
```

```python
df_feature2 = balanced_df.iloc [:, [0,2,5,7,8,10,11,16]]
```

We then split the data into test and training sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_feature2,  df_labels, test_size = 0.2, random_state = 0)
```

We use the random forest algorithm for prediction

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=0).fit(X_train,  y_train)
```

```python
Y_pred_rf = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
score_rf = round(accuracy_score(Y_pred_rf,y_test)*100,2)

print("The accuracy score achieved using RandomForestClassifier is: "+str(score_rf)+" %")
```
